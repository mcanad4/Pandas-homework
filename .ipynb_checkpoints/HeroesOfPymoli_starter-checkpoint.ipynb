```python
# Grouping the DataFrame by "SN"
player_group = purchase_data_df.groupby("SN")

# Count how many purchases each player had and create DataFrame
player_df = pd.DataFrame(player_group["Age Group"].value_counts())

# Rename the "SN" column to "Items Purchased"
player_df = player_df.rename(columns={"Age Group":"Items Purchased"})
player_df.reset_index()
player_df = player_df.sort_values("Items Purchased", ascending=False)
```

```python
#group and calculate total purchase prices
playergroup_df = purchase_data_df.groupby(['SN'])['Price'].sum().reset_index()
playergroup_df = playergroup_df.sort_values("Price", ascending=False)
```

```python
#merge with player_df 
player_complete = pd.merge(player_df, playergroup_df, how="left", on=["SN", "SN"])
avg_pur_price = player_complete["Price"]/player_complete["Items Purchased"]

#add average price to df
player_complete['Average Purchase Price']=avg_pur_price
player_complete = player_complete.sort_values("Price", ascending=False)
player_complete.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#group and calculate the sum of price
itemprice_df = purchase_data_df.groupby(['Item Name'])['Price'].sum().reset_index()
itemprice_df = itemprice_df.sort_values("Price", ascending=False)

#group and calculate count of items
itemcount_df = purchase_data_df.groupby(['Item Name'])['SN'].count().reset_index()
itemcount_df = itemcount_df.sort_values("SN", ascending=False)

#calculate item price
#item_price = itemprice_df["Price"]/itemcount_df["SN"]

#merge item dataframes
item_complete = pd.merge(itemprice_df, itemcount_df, how="left", on=["Item Name", "Item Name"])

#rename
item_complete = item_complete.rename(columns={"SN":"Purchase Count"})
item_complete = item_complete.rename(columns={"Price":"Total Purchase Value"})

#add item price to df
item_complete['Item Price']=item_complete['Total Purchase Value']/item_complete['Purchase Count']
item_complete = item_complete.sort_values("Purchase Count", ascending=False)

#format data
item_complete['Item Price'] = item_complete['Item Price'].map("${:.2f}".format)
item_complete['Total Purchase Value'] = item_complete['Total Purchase Value'].map("${:.2f}".format)
item_complete.head()

```
